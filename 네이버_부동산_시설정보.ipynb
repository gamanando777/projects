```python
# 관련 모듈 임포트
import pandas as pd
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import numpy as np
import time
from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.common.keys import keys
```

```python
# 매물번호 파일 불러오기
df = pd.read_csv('gu/df_gu_14.csv')
```

```python
# 1km이내 있는 시설이기때문에 거리 기본값 1001m

df['bus']=1001
df['subway']=1001
df['child']=1001
df['preschool']=1001
df['school']=1001
df['parking']=1001
df['market']=1001
df['convenience']=1001
df['laundry']=1001
df['bank']=1001
df['hospital']=1001
```

```python
# 2) 셀레니움 웹드라이버 객체 생성
# chromedriver.exe 위치 설정
driver = webdriver.Chrome('chromedriver.exe')

# 3) 웹브라우저 주소로 이동
# 지연시간 지정


none = []
idx=0
for li in df["apt"]:
    url = 'https://m.land.naver.com/complex/info/'+ str(li) + '?ptpNo=1'

    driver.get(url)
    time.sleep(2)

    # '시설'로 이동
    driver.find_element_by_xpath("//*[@id='_facil_trans_tab_cd']").click()
    time.sleep(2)

    # 시설 페이지 파싱
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    df.loc[idx,'title'] = soup.find('strong',class_='detail_complex_title').text
        
    # 주변 1km이내 편의시설 리스트
    com = soup.find_all('li',class_='basic_option_item')
        
#     time.sleep(2)    
    c=0    
    for c in range(len(com)):
            
        if com[c].text.strip().split()[0] == '버스정류장':
            df.loc[idx,'bus'] = com[c].text.strip().split()[1]
            
        elif com[c].text.strip().split()[0] == '지하철역':
            df.loc[idx,'subway'] = com[c].text.strip().split()[1]
            
        elif com[c].text.strip().split()[0] == '어린이집':
            df.loc[idx,'child'] = com[c].text.strip().split()[1]
            
        elif com[c].text.strip().split()[0] == '유치원':
            df.loc[idx,'preschool'] = com[c].text.strip().split()[1]
            
        elif com[c].text.strip().split()[0] == '학교':
            df.loc[idx,'school'] = com[c].text.strip().split()[1]
                
        elif com[c].text.strip().split()[0] == '주차장':
            df.loc[idx,'parking'] = com[c].text.strip().split()[1]
                
        elif com[c].text.strip().split()[0] == '마트':
            df.loc[idx,'market'] = com[c].text.strip().split()[1]
            
        elif com[c].text.strip().split()[0] == '편의점':
            df.loc[idx,'convenience'] = com[c].text.strip().split()[1]
            
        elif com[c].text.strip().split()[0] == '세탁소':
            df.loc[idx,'laundry'] = com[c].text.strip().split()[1]
                
        elif com[c].text.strip().split()[0] == '은행':
            df.loc[idx,'bank'] = com[c].text.strip().split()[1]
            
        elif com[c].text.strip().split()[0] == '병원':
            df.loc[idx,'bank'] = com[c].text.strip().split()[1]
                
        else:
            none.append(li)
            
#         time.sleep(2)
            
    idx += 1
    time.sleep(2)
    print("작동중")
```
